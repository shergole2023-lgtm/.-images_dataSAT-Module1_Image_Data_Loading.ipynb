## Setup and Imports

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
import glob

# Set up matplotlib for inline display
%matplotlib inline

# Define paths
base_path = './images_dataSAT/'
dir_non_agri = os.path.join(base_path, 'class_0_non_agri')
dir_agri = os.path.join(base_path, 'class_1_agri')

# Check if directories exist
print(f"Non-agriculture directory exists: {os.path.exists(dir_non_agri)}")
print(f"Agriculture directory exists: {os.path.exists(dir_agri)}")
```

## Task 1: Determine the shape (dimensions) of a single image

```python
# Task 1: Find shape of a single image

# Get all image files from non-agriculture directory
image_extensions = ['*.jpg', '*.jpeg', '*.png', '*.tif', '*.bmp']
image_files = []
for ext in image_extensions:
    image_files.extend(glob.glob(os.path.join(dir_non_agri, ext)))
    image_files.extend(glob.glob(os.path.join(dir_non_agri, ext.upper())))

if image_files:
    # Load the first image
    first_image_path = image_files[0]
    img = Image.open(first_image_path)
    
    # Convert to numpy array to get shape
    img_array = np.array(img)
    
    # Store in image_data variable (as per task)
    image_data = img_array
    
    print(f"Image path: {first_image_path}")
    print(f"Image shape: {image_data.shape}")
    print(f"Image mode: {img.mode}")
    
    # Display the image
    plt.figure(figsize=(6,6))
    plt.imshow(img)
    plt.title(f'Sample Image (Shape: {image_data.shape})')
    plt.axis('off')
    plt.show()
else:
    print("No images found in the directory")
    image_data = None
```

## Task 2: Display the first four images in ./images_dataSAT/class_0_non_agri directory

```python
# Task 2: Display first four non-agriculture images

# Get all image files from non-agriculture directory
non_agri_images = []
for ext in image_extensions:
    non_agri_images.extend(glob.glob(os.path.join(dir_non_agri, ext)))
    non_agri_images.extend(glob.glob(os.path.join(dir_non_agri, ext.upper())))

# Sort the images for consistent display
non_agri_images.sort()

print(f"Total non-agriculture images found: {len(non_agri_images)}")

# Display first 4 images
if len(non_agri_images) >= 4:
    fig, axes = plt.subplots(1, 4, figsize=(16, 4))
    
    for i in range(4):
        img_path = non_agri_images[i]
        img = Image.open(img_path)
        
        axes[i].imshow(img)
        axes[i].set_title(f'Non-Agri {i+1}\n{os.path.basename(img_path)}')
        axes[i].axis('off')
    
    plt.suptitle('First 4 Non-Agriculture Images', fontsize=16)
    plt.tight_layout()
    plt.show()
else:
    print(f"Less than 4 images found. Found {len(non_agri_images)} images.")
    # Display whatever is available
    if non_agri_images:
        fig, axes = plt.subplots(1, len(non_agri_images), figsize=(4*len(non_agri_images), 4))
        if len(non_agri_images) == 1:
            axes = [axes]
        for i, img_path in enumerate(non_agri_images):
            img = Image.open(img_path)
            axes[i].imshow(img)
            axes[i].set_title(f'Non-Agri {i+1}')
            axes[i].axis('off')
        plt.show()
```

## Task 3: Create a list named agri_images_paths with sorted full file paths

```python
# Task 3: Create sorted list of agriculture image paths

# Initialize empty list
agri_images_paths = []

# Get all image files from agriculture directory
for ext in image_extensions:
    agri_images_paths.extend(glob.glob(os.path.join(dir_agri, ext)))
    agri_images_paths.extend(glob.glob(os.path.join(dir_agri, ext.upper())))

# Sort the list (as required by the task)
agri_images_paths.sort()

# Display results
print(f"Number of agriculture images found: {len(agri_images_paths)}")
print("\nFirst 5 paths in the sorted list:")
for i, path in enumerate(agri_images_paths[:5]):
    print(f"{i+1}. {path}")

if len(agri_images_paths) > 5:
    print("...")
```

## Task 4: Determine the number of images of agricultural land

```python
# Task 4: Count agriculture images

# Count is simply the length of the list we created
num_agri_images = len(agri_images_paths)

print(f"=" * 50)
print(f"NUMBER OF AGRICULTURAL LAND IMAGES: {num_agri_images}")
print(f"=" * 50)

# Optional: Show file extensions distribution
extensions = {}
for path in agri_images_paths:
    ext = os.path.splitext(path)[1].lower()
    extensions[ext] = extensions.get(ext, 0) + 1

print("\nImage types distribution:")
for ext, count in extensions.items():
    print(f"  {ext}: {count} images")
```

## Task 5: Display the first four images of the agricultural land

```python
# Task 5: Display first four agriculture images

print(f"Displaying first 4 agriculture images out of {len(agri_images_paths)} total")

# Display first 4 images
if len(agri_images_paths) >= 4:
    fig, axes = plt.subplots(1, 4, figsize=(16, 4))
    
    for i in range(4):
        img_path = agri_images_paths[i]
        img = Image.open(img_path)
        
        axes[i].imshow(img)
        axes[i].set_title(f'Agriculture {i+1}\n{os.path.basename(img_path)}')
        axes[i].axis('off')
    
    plt.suptitle('First 4 Agriculture Images', fontsize=16)
    plt.tight_layout()
    plt.show()
    
    # Also display image information
    print("\nImage details:")
    for i in range(4):
        img_path = agri_images_paths[i]
        img = Image.open(img_path)
        img_array = np.array(img)
        print(f"Image {i+1}: {os.path.basename(img_path)} | Shape: {img_array.shape} | Mode: {img.mode}")
else:
    print(f"Less than 4 images found. Found {len(agri_images_paths)} images.")
    # Display whatever is available
    if agri_images_paths:
        fig, axes = plt.subplots(1, len(agri_images_paths), figsize=(4*len(agri_images_paths), 4))
        if len(agri_images_paths) == 1:
            axes = [axes]
        for i, img_path in enumerate(agri_images_paths):
            img = Image.open(img_path)
            axes[i].imshow(img)
            axes[i].set_title(f'Agriculture {i+1}')
            axes[i].axis('off')
        plt.show()
```
